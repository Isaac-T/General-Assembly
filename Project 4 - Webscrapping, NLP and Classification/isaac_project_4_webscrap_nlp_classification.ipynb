## Project 4 - Web Scrapping Data-related Jobs on mycareersfuture.

**Scenario**: 
I am working as a data scientist for a contracting firm that's rapidly expanding. They need to leverage data to win more contracts. The firm offers technology and scientific solutions and wants to be competitive in the hiring market. 

I have two main objectives:

   1. Determine the **industry factors** that are most important in **predicting the salary amounts** for these data.
   2. Determine the **factors that distinguish job categories and titles from each other**. For example, can required skills accurately predict job title?
   

   
To limit the scope, *focusing on data-related job postings*, e.g. data scientist, data analyst, research scientist, business intelligence in Singapore. I will be scraping my own data from mycareersfuture to best answer these two questions below.


### QUESTION 1: Factors that impact salary

I will be framing this as a classification problem, in which case i will create labels from salaries according to thresholds i set. I will identify features that impact **high or low salaries** jobs and identify features that lead to higher paying jobs.

### QUESTION 2: Factors that distinguish jobs

Using the job postings i have scraped, i will identify features in the data related to job postings that can distinguish **job titles** from each other. I am going to identify features that will differentiate data scientist from other jobs.


# Overview and Executive Summary

Dataset for this project was scrapped from https://www.mycareersfuture.sg/ on 2nd May 2019, where i filtered for 'Data' in the webpage and retrieved over 4700 job postings. 

From the original dataset scraped, i performed further filtering for jobs that were not/less related to data science and analytics. Thorough data cleaning was then conducted with the primary objective of removing null values and no data imputation was done. We eventually ended up with 1035 data science related job listings.

## Part 1 

The first half of this project will be focused on determining what factors would tentatively lead to higher paying jobs. 

The salary columns consist of salaries quoted in mixtures of monthly and annual terms. The first thing i did was to convert all the salary columns to monthly basis for direct comparison purpose. Through exploratory data analysis, i came to conclude that minimum salary will be set as our target variable instead of mean, median or maximum salaries(to be explained in detail later), with the threshold for higher paying jobs to be set as 7,000 dollars(75th percentile) monthly. 

I have also identified Job Seniority, Requirements and Responsibilities as my predictor variables.

Just through EDA, we are able to see that the below Seniority terms appeared more frequently in 'lower paying jobs'.
- Executive, Fresh/entry level, Senior executive, Junior Executive, Non-executive

and the respective Seniority terms appeared more often in 'higher paying jobs'.
- Manager, Middle Management, Senior Management

As for the job requirements and responsibilities text columns, i tried using WordCloud to identify requirements and responsibilities that seperates high and low paying jobs, however we could not identify distinct differences through EDA.

Next, i moved on to perform text data vectorization(counting words frequency) using 'CountVectorizer' and 'TfidfVectorizer'. Classification models explored in this project includes K-nearest neighbours, Logistic Regression and Random Forest. After trying several models and hyperparameters, i settled on using TfidfVectorizer to transform the text fields before running Logistic Regression as it gave us the best AUC and F1 score, balancing between both precision and accuracy scores.

**These are the key factors that seperates high and low paying jobs, they are listed in sequence with respect to feature importance.**

1) Seniority

- The main factor is Seniority. According to our model, Senior management tend to command higher pay. Whereas Fresh/entry level, Executive and Junior Executive had to settle for lower paying jobs.

2) Skills (Mainly Soft)

- Candidates with Leadership skills, having ability to come up with solutions for the organization, finance and technology knowledge, strong communication skills, management skills are able to command higher pay.

- Whereas Candidates with microsoft excel, analytics, visualization skillsets have lesser bargaining power with respect their pay.

3) Experience

- Candidates with more experience is able to command higher pay. For instance, according to our model, if you have a minimum of 10 years experience, you should fall in higher pay category (>$7,000 monthly)

## Part 2

The second half of the project is focused on identifying features that would differentiate/distinguish data scientist role from other data-related roles. The main objective is to identify what companies are looking for when they are hiring data scientist.

The first thing i did was to set data scientist job title as my target variable. Next, I identified job descriptions and requirements as my predictor variables.

Similar to part 1, i ran 'CountVectorizer' and 'TfidfVectorizer' as my word frequency counter. And explored K-nearest neighbours, Logistic Regression and Random Forest as my Classification models.

We were faced with an imbalanced classification problem, with data scientist only consisting of less than 10% of all our job postings. I oversampled our dataset using SMOTE(Synthetic Minority Over-sampling Technique) to adjust the class distribution.

After attempting the above classification models, i settled on Logistic Regression, but this time with CountVectorizer instead, as it gave us the best AUC and F1 score.

**These are the key factors that organizations look for when hiring data scientists.**

1) Skills

- Maching Learning, Python, Modeling, Mathematics

2) Experience

- Experience in the above skills

To conclude, if candidates want to land data scientist roles, they need to accumulate **Experience** in Machine Learning, Modeling and also Excel in both Python and Mathematics.

## Some cleaning needs to be done


```python
#Create a function to clean text strings for columns
def clean_text(string):

    clean_string = string.replace('\n','')\
                        .replace('\xa0','')\
                        .replace('&', 'and')
    return clean_string
```

```python
#Clean up the text in Responsibilities and Requirements
jobs_df_scraped['Responsibilities']=jobs_df_scraped['Responsibilities'].apply(clean_text)

jobs_df_scraped['Requirements']=jobs_df_scraped['Requirements'].astype(str)
jobs_df_scraped['Requirements']=jobs_df_scraped['Requirements'].apply(clean_text)
```

```python
#Create a set to retrieve the unique kinds of seniority
seniority = set()
for row in jobs_df_scraped['Seniority'].str.split(','):
    for rank in row:
        if rank not in seniority:
            seniority.add(' '.join(rank.split()))
```

```python
#Create a function to compare cross validated results of different models
def model_compare(models, X_train, y_train, score):
    scores_dict = {}
    
    for model in models:
        scores = cross_val_score(model,X_train,y_train,cv=10,verbose=1,n_jobs=-1, scoring=score)
        scores_dict[model]=scores
        
    return pd.DataFrame(scores_dict)
```

```python
#Create a function to return summary of classification metrics
def class_scorer(estimator,X,y):
    """
    INPUT: classification estimator, independent variables, target variable
    OUTPUT: confusion matrix, classification report
    """
    predictions = estimator.predict(X)
    pred_proba = estimator.predict_proba(X)
    print(confusion_matrix(y,predictions))
    print(classification_report(y,predictions))
    print()
    print('ROC AUC Score: ',roc_auc_score(y,pred_proba[:,1]))
```

```python
#Create a function to vectorize, split and retrieve feature names for text columns
def text_split(vectorizer,col):
    ''' 
    INPUT: vectorizer, column name
    OUTPUT: train set, test set, feature names
    '''
    train = vectorizer.fit_transform(jobs_df_scraped.loc[X_train.index][col])
    test = vectorizer.transform(jobs_df_scraped.loc[X_test.index][col])
    features = [col+': '+name for name in vectorizer.get_feature_names()]
    
    return train, test, features
```

```python
def logr_w_grid(X,y):
    """
    Gridsearch for Logistic Regression hyperparameters
    INPUT: train set of independent and target variable
    OUTPUT: the model fitted with the best estimator
    """
    
    logr = LogisticRegression()

    logr_params = {
        'penalty':['l1','l2'],
        'C':np.logspace(-3,2,30),
        'solver':['liblinear']
    }

    grid_logr = GridSearchCV(logr,param_grid=logr_params,cv=10,n_jobs=-1,verbose=1)
    grid_logr.fit(X,y)
    
    print(grid_logr.best_score_)
    print(grid_logr.best_params_)
    
    return grid_logr.best_estimator_
```

```python
optimal_logr = logr_w_grid(x3_train,y_train)
class_scorer(optimal_logr,x3_test,y_test)
```

```python
def rfc_w_grid(X,y):
    rfc = RandomForestClassifier()

    rfc_params = {
        'n_estimators':[300],
        'min_samples_split':range(2,8)
    }

    grid_rfc = GridSearchCV(rfc,param_grid=rfc_params,cv=10,n_jobs=-1,verbose=1)
    grid_rfc.fit(X,y)
    
    print('best score: ', grid_rfc.best_score_)
    print('best parameters: ', grid_rfc.best_params_)
    
    return grid_rfc.best_estimator_
```
