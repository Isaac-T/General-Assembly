```python
# KNN Imputation into LotFrontage

from sklearn.neighbors import KNeighborsRegressor 
impute_missing = house.loc[house['LotFrontage'].isnull(), :]
impute_valid = house.loc[~house['LotFrontage'].isnull(), :]

from sklearn.model_selection import cross_val_score

def find_best_k_reg(X, y, k_min=1, k_max=51, step=2, cv=10):
    k_range = range(k_min, k_max+1, step)
    r2s = []
    for k in k_range:
        knn = KNeighborsRegressor(n_neighbors=k)
        scores = cross_val_score(knn, X, y, cv=cv)
        r2s.append(np.mean(scores))
    return k_range[np.argmax(r2s)]

# Create impute_cols by excluding target variable, objects columns and missing columns
missing_cols=[column for column in house if house[column].isnull().any()]
category_columns = [column for column in house if house[column].dtypes == object]
impute_cols = [c for c in impute_valid.columns if not c in ['SalePrice'] + missing_cols + category_columns]

# x-true & y-true
y = impute_valid.LotFrontage.values
X = impute_valid[impute_cols]

# set and fit the scaler
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
Xs = ss.fit_transform(X)

# call/use find k-best function on known data
best_k = find_best_k_reg(Xs, y)

# apply k-best to fit model
knn = KNeighborsRegressor(n_neighbors=best_k)
knn.fit(Xs, y)

# prepair rows with missing target values
X_miss = impute_missing[impute_cols]
X_miss_s = ss.transform(X_miss)

# use model to predict unknown values
LotFrontage_impute = knn.predict(X_miss_s)

house_imputed = house.copy()

house_imputed.loc[house['LotFrontage'].isnull(), 'LotFrontage'] = LotFrontage_impute
```

# Dummify Categorical Columns

Creating a function that drops the MOST COMMON OCCURING element instead of the first element

```python
def get_dummies(df): 
    '''
    DOCSTRING: This function takes in a DataFrame and converts all columns into dummy variables. 
               The element that occurs the most will be regarded as the default for hot-one encoding and dropped. 
    INPUT: DataFrame
    OUTPUT: DataFrame
    '''
    #Create an empty DataFrame to append to
    dummy_cat_col = pd.DataFrame()

    #iterate through each column in df
    for col in df.columns:
        #create dummy variables from the column
        new_col = pd.get_dummies(df[col])
        #drop the column that was the most common occurring element in the original column (becomes the default for hot-one encoding)
        new_col.drop(df[col].value_counts().first_valid_index(),axis=1,inplace=True)
        #rename the columns with original column name included
        new_col.columns=[col+'_'+ str(name) for name in new_col.columns]
        #append to the list of dummy variables 
        dummy_cat_col = pd.concat([dummy_cat_col,new_col],axis=1)

    return dummy_cat_col
```

# Lasso Regularization

```python
from sklearn.linear_model import Lasso, LassoCV

# Examine BEST alpha coefficients by testing with 200 different alpha values   
optimal_lasso = LassoCV(n_alphas=200, cv=10, verbose=1)
optimal_lasso.fit(Xs_rfecv_train, y_train)
optimal_lasso.alpha_
print('best alpha: {}'.format(optimal_lasso.alpha_))
print(optimal_lasso.coef_)

# Cross-validate the lasso R2 with the optimal alpha.
lasso = Lasso(alpha=optimal_lasso.alpha_)
lasso_scores = cross_val_score(lasso, Xs_rfecv_train, y_train, cv=10)
print(lasso_scores)
print('mean score: {}'.format(np.mean(lasso_scores)))
print('std: {}'.format(np.std(lasso_scores)))
```

# Ridge Regularization

```python
from sklearn.linear_model import Ridge, RidgeCV

## Finding the BEST alpha value & best error
# Recall that ridge performs best when searching alphas through logarithmic space (np.logspace).
ridge_alphas = np.logspace(-2, 7, 50)

# select the best alpha with RidgeCV
optimal_ridge = RidgeCV(alphas=ridge_alphas, cv=10)
optimal_ridge.fit(Xs_rfecv_train, y_train)
print('best alpha: {}'.format(optimal_ridge.alpha_))

# Cross-validate the ridge regression R2 with the optimal alpha.
ridge = Ridge(alpha=optimal_ridge.alpha_)
ridge_scores = cross_val_score(ridge, Xs_rfecv_train, y_train, cv=10)

print(ridge_scores)
print('mean score: {}'.format(np.mean(ridge_scores)))
print('std: {}'.format(np.std(ridge_scores)))
```
